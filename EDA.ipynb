## EDA

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

##### Configuración de estilo 

```python
sns.set(style="whitegrid")
```

##### Leer el archivo CSV

```python
df = pd.read_csv('DATOSFINALESFILTRADOS.csv')
```

##### Inspeccionar los datos

```python
print("Información del dataset:")
print(df.info())
print("\nDescripción del dataset:")
print(df.describe())
```

##### Analizar la distribución de las variables
##### Distribución de la variable 'Survived'

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', data=df)
plt.title('Distribución de Supervivencia')
plt.show()
```

##### Distribución de la variable 'Pclass'

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', data=df)
plt.title('Distribución de Clases')
plt.show()
```

##### Distribución de la variable 'Sex'

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', data=df)
plt.title('Distribución por Sexo')
plt.show()
```

##### Distribución de la variable 'Age'

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'].dropna(), kde=True, bins=30)
plt.title('Distribución de Edad')
plt.show()
```

##### Distribución de la variable 'Fare'

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Fare'].dropna(), kde=True, bins=30)
plt.title('Distribución de Tarifas')
plt.show()
```

##### Analizar las relaciones entre las variables
##### Supervivencia por clase

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Supervivencia por Clase')
plt.show()
```

##### Supervivencia por sexo

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Supervivencia por Sexo')
plt.show()
```

##### Supervivencia por edad

```python
plt.figure(figsize=(8, 6))
sns.histplot(df[df['Survived'] == 1]['Age'].dropna(), kde=True, bins=30, color='green', label='Survived')
sns.histplot(df[df['Survived'] == 0]['Age'].dropna(), kde=True, bins=30, color='red', label='Not Survived')
plt.title('Supervivencia por Edad')
plt.legend()
plt.show()
```

##### Supervivencia por tarifa

```python
plt.figure(figsize=(8, 6))
sns.histplot(df[df['Survived'] == 1]['Fare'].dropna(), kde=True, bins=30, color='green', label='Survived')
sns.histplot(df[df['Survived'] == 0]['Fare'].dropna(), kde=True, bins=30, color='red', label='Not Survived')
plt.title('Supervivencia por Tarifa')
plt.legend()
```

##### Supervivencia por puerto de embarque

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Embarked', hue='Survived', data=df)
plt.title('Supervivencia por Puerto de Embarque')
plt.show()
```
